```
import pandas as pd
import re

DF = pd.read_excel('Data/database.xlsx', 'Calisthenics')

def func(cell):
  cell = [elem for elem in str(cell).split('-')]
  return(cell)

DF[['Rests', 'Weights', 'Reps']] = DF[['Rests', 'Weights', 'Reps']].applymap(lambda cell: func(cell))

def func_weights(row):
  weights = row.Weights
  if len(weights)==1:
    weights = [float(weights[0])]*len(row.Reps)
  elif len(weights)==len(row.Reps):
    weights = [float(weight) for weight in row.Weights]
  else:
    weights = 'ERROR'
  return(weights)
DF['Weights'] = DF.apply(lambda row: func_weights(row), axis=1)

def func_rests(row):
  rests = row.Rests
  if len(rests)==1:
    rest = rests[0]
    if re.match('\d+\*', rest) is None:
      rests = [float(rest)]*len(row.Reps)
    else:
      rest = re.sub('\*', '', rest)
      rests = [float(rest)*float(rep) for rep in row.Reps]
  elif len(rests)==len(row.Reps):
    rests = [float(rest) for rest in row.Rests]
  else:
    rests = 'ERROR'
  return(rests)
DF['Rests'] = DF.apply(lambda row: func_rests(row), axis=1)

def func_workout(row):
  workout = ['{}x{} ({}s)'.format(row.Reps[k], row.Weights[k], row.Rests[k]) for k in range(len(row.Reps))]
  workout = ' - '.join(workout)
  return(workout)
DF['Workout'] = DF.apply(lambda row: func_workout(row), axis=1)

display(DF.tail())
```
